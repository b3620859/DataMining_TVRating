```python
def create_date(datestart = None,dateend = None):
    global date_list #設置全局變量
    # 創建日期表
    if datestart is None:
        datestart = '20190402'
    if dateend is None:
        dateend = datetime.datetime.now().strftime('%Y%m%d')

    # 轉換日期格式
    datestart=datetime.datetime.strptime(datestart,'%Y%m%d')
    dateend=datetime.datetime.strptime(dateend,'%Y%m%d')
    date_list = []
    date_list.append(datestart.strftime('%Y%m%d'))
    while datestart<dateend:
    # 日期疊加一日
        datestart+=datetime.timedelta(days=+1)
    # 將日期由字串轉為列表
        date_list.append(datestart.strftime('%Y%m%d'))
#     return date_list
    print(date_list)

    
# def get_data(date_list = None):
#     for date in date_list:
#         date = int(date) - 20000000
# #         print(date, type(date))
#         src = 'http://www1.xkm.com.tw/hr/DATA/HR'+str(date)+'.htm'
# #         print(src)
#         src = src = requests.get('http://www1.xkm.com.tw/hr/DATA/HR' +str(date)+'.htm')
#         src_text = src.text
# #         print(src_text)
#         dfs = pd.read_html('http://www1.xkm.com.tw/hr/DATA/HR' +str(date)+'.htm')
#         rating = dfs[0]
#         rating.columns = rating.loc[3]
#         rating.drop([0,1,2,3,124,125,126],inplace = True)
#         rating.reset_index(drop = True, inplace=True)
#         rating.to_csv(str(date)+".csv" ,index=False,encoding = 'utf-8-sig')

        
datestart = '20190331'
dateend = '20190331'
create_date(datestart,dateend)
# if __name__ == '__main__':
#     datestart = '20190321'
#     dateend = '20190331'
#     create_date(datestart,dateend)
#     get_data(date_list)
```
